D) Run time

F) Votes    

```python
# Url = https://www.imdb.com/list/ls095964455/ --this url got 404 not found
```

```python
driver.get('https://www.imdb.com/')
```

```python
menu = driver.find_element(By.XPATH,'/html/body/div[2]/nav/div[2]/label[1]/span')
menu.click()
```

```python
search = driver.find_element(By.XPATH,'/html/body/div[2]/nav/div[2]/aside[1]/div/div[2]/div/div[2]/div[1]/span/div/div/ul/a[3]/span')
search.click()
```

```python
tv_name = []
tv_year = []
run_time = []
tv_ratings = []
tv_votes = []
```

```python
try:
    for name in driver.find_elements(By.XPATH,'//div[@class="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b189961a-9 iALATN cli-title"]/a/h3')[:97]:
        tv_name.append(name.text)
except:
    print("Not Found")
    
tv_name 
```

```python
try:
    for year in driver.find_elements(By.XPATH,'//div[@class="sc-b189961a-7 feoqjK cli-title-metadata"]/span[1]')[:97]:
        tv_year.append(year.text)
except:
    print("Not Found")  
    
tv_year   
```

```python
try:
    for episode in driver.find_elements(By.XPATH,'//div[@class="sc-b189961a-7 feoqjK cli-title-metadata"]/span[2]')[:97]:
        run_time.append(episode.text)
except:
    print("Not Found")    
    
run_time   
```

```python
try:
    for rate in driver.find_elements(By.XPATH,'//span[@class="sc-b189961a-1 kcfvgk"]/div/span[1]')[:97]:
        tv_ratings.append(rate.text.replace('\n','').split(' ')[0])
except:
    print("Not Found")   
 

tv_ratings
```

```python
try:
    for vote in driver.find_elements(By.XPATH,'//span[@class="ipc-rating-star--voteCount"]')[:97]:
        if vote.text is None:
            tv_votes.append('-')
        else:
            tv_votes.append(vote.text)
except:
    print("Not Found")
    
tv_votes   
```

```python
len(tv_name),len(tv_ratings),len(tv_votes),len(tv_year),len(run_time)
```

```python
IMDB = pd.DataFrame({
    'TV Show Name':tv_name,
    'Year':tv_year,
    'Episodes':run_time,
    'Rating':tv_ratings,
    'Votes':tv_votes
})

IMDB
```

```python
IMDB.to_csv('IMDB_dataset.csv')
```
